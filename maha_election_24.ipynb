```python
import pandas as pd
col_name = ['Constituency','CandidateName','Party','Result','Votes_Status']
df = pd.read_csv('Data/maha-election-results.csv',delimiter = '|', names = col_name)
```

```python
df
```

```python
df.head(10)
```

# Make new columns for votes and margin

```python
df['Votes_Status']
```

```python
ser_1 = df['Votes_Status'].str.split("(")
```

```python
ser_1
```

```python
votes = ser_1.apply(lambda x:x[0])
df['Votes'] = votes
```

```python
df.head(10)
```

```python
votes_margin = ser_1.apply(lambda x: x[1])
df['Votes_Margin'] = votes_margin
```

```python
df.head(5)
```

```python
df['Votes_Margin'] = df['Votes_Margin'].str.replace(" ","")
```

```python
df['Votes_Margin'] = df['Votes_Margin'].str.replace(")","")
```

```python
df
```

```python
df.info()
```

# convert votes columns to numeric

```python
df['Votes'] = df['Votes'].astype(int)
df['Votes_Margin'] = df['Votes_Margin'].astype(int)
```

```python
df.head(5)
```
